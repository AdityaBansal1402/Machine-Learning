```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist)
print (dist.shape)
```

```python
plt.style.use("seaborn")
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
cov=np.array([[1,0.5],[0.5,1]])
hehe=np.random.multivariate_normal(mean,cov,500)
print(hehe)
```

```python
plt.scatter(hehe[:,0],hehe[:,1])
plt.show()
```

```python
mean1=np.array([5.0,6.0])
cov1=np.array([[1.3,0.2],[0.2,1.1]])
dist=np.random.multivariate_normal(mean,cov,500)
dist1=np.random.multivariate_normal(mean1,cov1,500)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist1[:,0],dist1[:,1])
plt.show()
```
