```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate a Normal distribution (1-D)
plt.style.use("seaborn-dark")
u=5;
sigma=2;
vals=u+sigma*np.random.randn(100)
print(vals.shape)
print(vals)
plt.hist(vals,10)
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

# MULTIVARIATE NORMAL DISTRIBUTION

```python
#BiVariate
mean=np.array([1.0,1.0])
cov=np.array([[20,8],[8,20]])
mean1=np.array([25.0,25.0])
cov1=np.array([[25,2],[2,25]])
dist=np.random.multivariate_normal(mean,cov,500)
dist1=np.random.multivariate_normal(mean1,cov1,500)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist1[:,0],dist1[:,1])
plt.show()
```
