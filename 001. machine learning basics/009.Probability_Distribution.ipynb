```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate a Normal distribution (1-D)
plt.style.use("seaborn-dark")
u=5;
sigma=2;
vals=u+sigma*np.random.randn(100)
print(vals.shape)
print(vals)
plt.hist(vals,10)
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```
