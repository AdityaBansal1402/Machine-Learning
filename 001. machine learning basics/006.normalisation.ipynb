```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('seaborn')
dfx=pd.read_csv('./000.datasets/003.linearX.csv')
dfy=pd.read_csv('./000.datasets/004.linearY.csv')
x=dfx.values
y=dfy.values
plt.scatter(x,y)
plt.show()
```

```python
#normalisation
x2=(x-x.mean())/x.std()
y2=(y-y.mean())/y.std()
plt.scatter(x,y,color='green',label='Raw Data')
plt.scatter(x2,y2,color='orange',label='Normalised Data')
plt.legend()
plt.show()
```
