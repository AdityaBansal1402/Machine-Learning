```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn-dark")
```

```python
#LINE GRAPH
plt.style.use("dark_background")
plt.figure(figsize=(10,10))
x=np.arange(10)
y1=x**2
y2=x*2+3
plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.plot(x,y2,color='green',label="Kiwi",marker='^',linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.ylim(0,100)
plt.xlim(0,10)
plt.title("Price of fruits over time")
plt.legend()
plt.show()
```

```python
#SCATTER PLOTS
plt.style.use("dark_background")
plt.figure(figsize=(5,5))
x=np.arange(10)
y1=x**2
y2=x*2+3
plt.scatter(x,y1,color='red',label="Apple",marker='o')
plt.scatter(x,y2,color='green',label="Kiwi",marker='^',linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.ylim(0,100)
plt.xlim(0,10)
plt.title("Price of fruits over time")
plt.legend()
plt.show()
```

```python
#BAR GRAPH
plt.figure(figsize=(10,10))
a=np.array([0,1,2])*2
plt.bar(a,[10,20,15],width=0.5,tick_label=["     Gold","     Steel","     Tin"],label="2021",color="orange")
plt.bar(a+0.5,[20,40,10],width=0.5,label="2022",color="skyblue")
plt.xlabel("Metals")
plt.ylabel("sale in cr")
plt.ylim(0,50)
plt.xlim(-1,6)
plt.title("Sales of metals in 2021 and 2022")
plt.legend()
plt.show()
```

```python
#PIE CHART
plt.figure(figsize=(10,10))
subjects="Maths","Chem","Physics","English"
weightage=[20,10,15,5]
plt.pie(weightage,labels=subjects,explode=(0,0,0.2,0),autopct='%1.1f%%')
plt.show()
```

```python
#HISTOGRAM
plt.style.use('seaborn')
xsn=np.random.randn(100)
sigma=5
u=70
x=np.round(xsn*sigma+u)
print(x)
```

```python
y=np.round(xsn*sigma+(u-20))
plt.hist(x)
plt.hist(y,alpha=0.9)
plt.show()
```
